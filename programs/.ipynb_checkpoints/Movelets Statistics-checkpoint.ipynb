```python
import json
import glob
import os
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# DIR = '/home/tarlis/Research/tarlis/results/'
DIR = './../results/'
```

```python
def redefine_dataframe(df):
#     names = df.columns.tolist()
#     new = []
#     names.remove('tid')
#     names.remove('label')
#     names.remove('size')
#     names.remove('quality')
#     names.remove('movelet_id')
#     print(names)
#     for x in names:
#         new.append(x.split('.')[0])
#     new.append('tid')
#     new.append('label')
#     new.append('size')
#     new.append('quality')
#     new.append('movelet_id')
#     df.columns = new
    df = df.fillna('-')
    return df
```

```python
def movelets_dataframe(path_name):
    count = 0
    path_to_file = glob.glob(os.path.join(path_name, '**', 'moveletsOnTrain.json'), recursive=True)
    df = pd.DataFrame()
    for path in path_to_file:
        with open(path) as f:
            data = json.load(f)    
        l = len(data['shapelets'])
        for x in range(0, l):
            aux_df = []
            
            points = data['shapelets'][x]['points_with_only_the_used_features']
            aux_df = json_normalize(points)
            
            aux_df['tid'] = data['shapelets'][x]['trajectory']
            aux_df['label'] = data['shapelets'][x]['label']
            aux_df['size'] = int(data['shapelets'][x]['quality']['size'])
            aux_df['quality'] = int(data['shapelets'][x]['quality']['quality'] * 100)
            aux_df['movelet_id'] = count
            df = df.append(aux_df)
            count += 1
#     print(df)
    return redefine_dataframe(df)
## remover iguais?
def candidates_dataframe(path_name):
    count = 0
    path_to_file = glob.glob(os.path.join(path_name, '**', 'candidatesOnTrain.json'), recursive=True)
    df = pd.DataFrame()
    for path in path_to_file:
        with open(path) as f:
            data = json.load(f)    
        l = len(data['shapelets'])
        for x in range(0, l):
            aux_df = []
            
            points = data['shapelets'][x]['points_with_only_the_used_features']
            aux_df = json_normalize(points)
            
            aux_df['tid'] = data['shapelets'][x]['trajectory']
            aux_df['label'] = data['shapelets'][x]['label']
#             aux_df['size'] = int(data['shapelets'][x]['quality']['size'])
#             aux_df['quality'] = int(data['shapelets'][x]['quality']['quality'] * 100)
            aux_df['proportion'] = int(data['shapelets'][x]['quality']['proportion'])
            aux_df['movelet_id'] = count
            df = df.append(aux_df)
            count += 1
#     print(df)
    return redefine_dataframe(df)
```

```python
prefix = 'SMLD-5fold_4T_60G/'
method = 'Hiper_MM-T2'
# -------------------------

path_name = DIR + prefix + method + '/'
df = movelets_dataframe(path_name)
```
